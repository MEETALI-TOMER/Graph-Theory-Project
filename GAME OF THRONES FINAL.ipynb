```python
import pandas as pd
import matplotlib.pyplot as plt

book1 = pd.read_csv('C:/Users/Meetali Tomer/Downloads/datasets/book1.csv')

book1.head()
```

```python
import networkx as nx
G_book1 = nx.Graph()

for _, edge in book1.iterrows(): 
    G_book1.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])

books = [G_book1]
book_fnames = ['C:/Users/Meetali Tomer/Downloads/datasets/book2.csv', 'C:/Users/Meetali Tomer/Downloads/datasets/book3.csv', 'C:/Users/Meetali Tomer/Downloads/datasets/book4.csv', 'C:/Users/Meetali Tomer/Downloads/datasets/book5.csv']
for book_fname in book_fnames:
    book = pd.read_csv(book_fname)
    G_book = nx.Graph()
    for _, edge in book.iterrows():
        G_book.add_edge(edge['Source'], edge['Target'], weight=edge['weight'])
    books.append(G_book)
```

```python
i=1
for book in books:
    print("Book" , i ,len(book.nodes()))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i ,len(book.edges()))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i , nx.average_shortest_path_length(book))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i , nx.diameter(book))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i , nx.average_clustering(book))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i , nx.transitivity(book))
    i+=1
```

```python
i=1
for book in books:
    print("Book" , i, nx.density(book))
    i+=1
```

```python
%matplotlib inline

evol = [nx.degree_centrality(book) for book in books]

degree_evol_df = pd.DataFrame.from_records(evol)

set_of_char = set()
for i in range(5):
    set_of_char |= set(list(degree_evol_df.T[i].sort_values(ascending=False)[0:4].index))
list_of_char = list(set_of_char)

degree_evol_df[list_of_char].plot(figsize=(13, 7))
```
